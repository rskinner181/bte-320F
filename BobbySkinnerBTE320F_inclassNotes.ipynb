bubble set
--> gets a number and tries it with each in list until it is sorted

```python
def quicksort(L):
  if len(L) < 2:
    return L
  else:
    pivot = L[len(L)//2]
    less, equal, greater = [], [], []

    for item in L:
      if item < pivot:
        less.append(item)
      elif item == pivot:
        equal.append(item)
      else:
        greater.append(item)
    return quicksort(less) + equal + quicksort(greater)
quicksort([8,2,5,1,6,9,7])
```

```python
from typing import NewType
class ATM:
  def __init__(self, balance):
    self.__balance = balance
    self.__type = "checking"

  def get_type(self):
    return self.__type

  def set_type(self, newType):
    self.__type = NewType

  def get_balance(self):
    return self.__balance



  def deposit(self, amount):
    self.__balance += amount
    print(f"Your new balance: {self.__balance}$")

  def withdraw(self, amount):
    if self.__balance < amount:
      print("Insufficient funds...")
    else:
      self.__balance -= amount
      print(f"Your new balance: {self.__balance}$")
```

```python
from numba import njit

@njit

def monte_carlo_pi(N):
  acc = 0
  for i in range(N):
    x = random.random()
    y = random.random()
    if x ** 2 + y ** 2 <= 1.0:
      acc += 1
  return 4 * acc / N

%timeit monte_carlo_pi(100_000)
```
