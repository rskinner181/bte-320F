```python
sales = float(input("Enter sales ($): "))

cogs = 0.4 * sales
gross_profit = sales - cogs
tax_rate = input("Enter Tax Rate (%): ")
tax_rate = float(tax_rate)
tax = gross_profit * (tax_rate/100)
net_income = gross_profit - tax

print(f"Gross Profit: ${gross_profit}")
print(f"Tax: ${tax}")
print(f"Net Income: ${net_income}")
print(f"Sales: ${sales}")
print(f"COGS: ${cogs}")
```

In class assignment 2 -- numerical calculator

```python
n1 = float(input("Enter first number: "))
n2 = float(input("Enter second number: "))
op = input("Enter operator: ")

if op == "+":
    result = n1 + n2
    print(f"{n1} + {n2} = {result}")

elif op == "-":
    result = n1 - n2
    print(f"{n1} - {n2} = {result}")

elif op == "*":
    result = n1 * n2
    print(f"{n1} * {n2} = {result}")

elif op == "/":
    result = n1 / n2
    print(f"{n1} / {n2} = {result}")

else:
    print("Invalid operator")
    result = None
```

```python
def total(hours):
    flat_fee = 5
    hourly_rate = 2.5
    total = flat_fee + hourly_rate * hours
    return total

print("Parking Fee Calculation (1-8 hours):")
for h in range(1, 9):
    fee = total(h)
    print(f"For {h} hour(s): ${fee:.2f}")
```

```python
print("Hour(s)", "Fee")
for hour in range (1,9):
    fee = 5 + 2.5 * hour
    if fee < 10.0:
        fee = 10.0
    elif fee > 20.0:
        fee = 20.0
    print(f"{hour}  ${fee}")
```

```python
import math

ticketprice = 10
standardAttendance = 20

print("Ad$\tProfit")
for ad in range (0,201,25):
  revenue = (20 + 2 * round(math.sqrt(ad)) ** 0.5) * ticketprice
  costs = 200 + ad
  profit = revenue - costs
  print(f"{ad}\t $  {profit}")
```

For and while loops

```python
print("Hour(s)", "Fee")
for hour in range (1,9):
    fee = 5 + 2.5 * hour
    if fee < 10.0:
        fee = 10.0
    elif fee > 20.0:
        fee = 20.0
    print(f"{hour}  ${fee}")
```

```python
print("Hour(s)", "Fee")
hour = 1
while hour < 9:
    fee = 5 + 2.5 * hour
    if fee < 10.0:
        fee = 10.0
    elif fee > 20.0:
        fee = 20.0
    print(f"{hour}  ${fee}")
    hour += 1
```
