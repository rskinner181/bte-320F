###Rock Paper Scissors

```python
import random

choices = ['rock', 'paper', 'scissors']

play = input("Do you want to play a game (y/n)? ")

while play == "y":
    user_move = input("Enter rock, paper, or scissors: ").lower()

    if user_move not in choices:
        print("Invalid choice, try again.")
        continue

    computer_move = random.choice(choices)
    print(f"The computer chose: {computer_move}")

    if computer_move == user_move:
        print("It's a tie")
    elif (computer_move == "scissors" and user_move == "paper") or \
         (computer_move == "paper" and user_move == "rock") or \
         (computer_move == "rock" and user_move == "scissors"):
        print("You lose")
    else:
        print("You win")

    play = input("Do you want to play again (y/n)? ")

print("Thanks for playing!")
```

###ROI Calculator

```python
def roi(cost: float, revenue: float) -> float:
    assert cost > 0, "Cost must be positive"
    assert isinstance(cost, (int, float)), "Cost must be numerical"
    assert isinstance(revenue, (int, float)), "Revenue must be numerical"

    return (revenue - cost) / cost

if __name__ == "__main__":
    c = float(input('Enter cost:'))
    r = float(input('Enter revenue:'))
    ret = roi(c, r)
    print(ret)
```

###Loan Payment Calculator

```python
def loan_payment(principal: float, annual_rate: float, months: int) -> float:
    assert isinstance(principal, (int, float)), "principal must be numeric"
    assert isinstance(annual_rate, (int, float)), "annual_rate must be numeric"
    assert isinstance(months, int), "months must be an integer"

    if principal <= 0:
        raise ValueError("Principal must be > 0.")
    if annual_rate < 0:
        raise ValueError("Annual rate cannot be negative.")
    if months <= 0:
        raise ValueError("Number of months must be > 0.")

    r = annual_rate / 12.0
    n = months
    P = float(principal)

    if r == 0.0:
        return P / n
    else:
        return r * P / (1.0 - (1.0 + r) ** (-n))

if __name__ == "__main__":
    try:
        P = float(input("Enter principal: "))
        apr = float(input("Enter annual interest rate as a decimal: "))
        n = int(input("Enter number of months: "))
        payment = loan_payment(P, apr, n)
        print(f"Monthly payment: ${payment:.2f}")
    except Exception as e:
        print(e)
```

###Quality Control Manager (Exceptions)

```python
def getValues(n):
    assert isinstance(n, int), "n must be an integer"
    assert n >= 0, "n must be a positive integer"
    items = []
    for i in range(n):
        items.append(float(input(f"Enter value: ")))
    return items

def calcAverage(alist):
    assert isinstance(alist, list), "Input parameter 'alist' must be a list."

    subset = alist[3:]

    if len(subset) == 0:
        raise ValueError("The length of 'alist' must be greater than 3.")

    return sum(subset) / len(subset)

def main():
    try:
        n = int(input("Enter number of items: to be added to the list: "))
        alist = getValues(n)
        avg = calcAverage(alist)
        print(f"Average: {avg}")
    except ValueError as e:
        print(f"Error: {e}")
    except Exception as e:
        print(f"Error: {e}")

main()
```

###Greeting Class

```python
class Greeting:
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name = last_name

    def display(self, formal):
        if formal:
            print(f"{self.last_name}, {self.first_name}")
        else:
            print(f"{self.first_name}")

    def __str__(self):
        return f"{self.first_name} {self.last_name}"

person = Greeting("Jane", "Doe")

person.display(formal=True)
person.display(formal=False)
print(person)
```

###Student Class

```python
class Student:
    # 1. Added 'courses' as a parameter here
    def __init__(self, name, number, courses=[]):
        self.__name = name
        self.__number = number
        # 2. This assigns the parameter to the instance attribute
        self.__courses = courses

    def get_name(self):
        assert isinstance(self.__name, str), "Name must be a string."
        return self.__name

    def get_courses(self):
        return self.__courses

    def set_name(self, newName): # Renamed for clarity
        self.__name = newName

    def __len__(self):
        return len(self.__courses)

    def enroll(self, newCourse):
        if newCourse not in self.__courses:
            self.__courses.append(newCourse)
        else:
            print(f"Student {self.__name} is already enrolled in {newCourse}.")
```

```python
s = Student("Jane Doe", 12345)
print(f"Student Name: {s.get_name()}")
print(f"Classes already enrolled in: {s.get_courses()}")
s.enroll('BTE 320')
print(f"Classes already enrolled in: {s.get_courses()}")
s.enroll('BTE 320')
print(f"Total number of classes enrolled: {len(s)}")
```

###Classes Point and Rectangle

```python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def distance(self, other):
        dx = (self.x - other.x) ** 2
        dy = (self.y - other.y) ** 2
        return (dx + dy) ** 0.5

class Rectangle:
    def __init__(self, p1, p2, p3, p4):
        self.p1 = p1
        self.p2 = p2
        self.p3 = p3
        self.p4 = p4

        self.width = self.p1.distance(self.p2)
        self.height = self.p1.distance(self.p3)

    def area(self):
        return self.width * self.height

p1 = Point(0, 0)
p2 = Point(0, 6)
p3 = Point(6, 0)
p4 = Point(6, 6)

r = Rectangle(p1, p2, p3, p4)
print(f"Area: {r.area()}")
```

###Class Employee - Manager

```python
class Employee:
    def __init__(self, name, wage, hours):
        self.name = name
        self._hourly_wage = wage
        self.hours = hours

    def get_wage(self):
        return self._hourly_wage

    def set_hourly_wage(self, new_wage):
        if new_wage < 0:
            print("Error: Hourly wage cannot be negative. Wage not updated.")
        else:
            self._hourly_wage = new_wage
            print(f"Hourly wage updated to: ${self._hourly_wage:.2f}")

    def calculate_pay(self):
        return self._hourly_wage * self.hours

    def display_info(self):
        print(f"Employee Name: {self.name}")
        print(f"Hourly Wage: ${self._hourly_wage:.2f}")
        print(f"Hours Worked: {self.hours}")
        print(f"Total Pay: ${self.calculate_pay():.2f}")

class Manager(Employee):
    def __init__(self, name, wage, hours, bonus):
        super().__init__(name, wage, hours)
        self.bonus = bonus

    def calculate_pay(self):
        return (1 + self.bonus) * self.get_wage() * self.hours

employee = Employee("Bob", 50, 10)
manager = Manager("Alice", 100, 15, 0.2)

print(f"Employee {employee.name} weekly payment is ${employee.calculate_pay():.2f}")
print(f"Manager {manager.name} weekly payment is ${manager.calculate_pay():.2f}")
```
